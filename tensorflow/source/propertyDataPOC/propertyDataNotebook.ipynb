```python
categoricalColumns = ['propertyType', 'oldOrNew', 'duration', 'locality', 'city', 'district', 'county', 'postcode', 'street']
aggregations = ['count', 'sum', np.mean, 'std', min, max]
```

```python
# Group by categorical data.
grouped = {}
for c in categoricalColumns:
    grouped[c] = df[['price',c]].groupby(c).agg(aggregations)['price'].sort_values(by='mean', ascending=False)
```
